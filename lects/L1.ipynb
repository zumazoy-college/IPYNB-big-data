```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import desc
```

```python
spark = SparkSession.builder.appName('IceCream').getOrCreate()
```

```python
df = spark.read.csv('IceCream_data.csv', header=True, inferSchema=True)
```

```python
ice_cream_country = df.groupBy('страна').sum('количество_мороженного')
print(ice_cream_country.orderBy(desc('sum(количество_мороженного)')).limit(10).show())
```

```python
spark.stop()
```
