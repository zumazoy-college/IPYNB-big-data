<h5>Ход работы:</h5>

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import desc, avg, count, max, min, col
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
spark = SparkSession.builder.appName('CarDataAnalysis').getOrCreate() # Создание Spark сессии

df = spark.read.csv('cardekho.csv', header=True, inferSchema=True) # Загрузка данных
```

```python
# Просмотр структуры данных
df.printSchema()

# Количество строк и столбцов
print(f"Количество строк: {df.count()}")
print(f"Количество столбцов: {len(df.columns)}")

# Просмотр первых строк
df.show(5, truncate=False)

# Статистика по числовым столбцам
df.describe().show()

# Базовый анализ: группировка по топливу
fuel_count = df.groupBy('fuel').count()
fuel_count.show()

# Анализ: средняя цена по типу топлива
avg_price_by_fuel = df.groupBy('fuel').agg(avg('selling_price').alias('avg_price'))
avg_price_by_fuel.orderBy(desc('avg_price')).show()
```

```python
spark.stop()
```
